# find_qa_dataset: Scans dataset names for QA/quality keywords

extract_cloud_mask:

    Opens each HDF, selects the QA layer,

    Converts it into a binary cloud mask (cloud vs. clear),

    Writes out a new HDF containing only that mask.

Loop at the bottom: Processes every .hdf in Drive folder

```python
from pyhdf.SD import SD, SDC
import numpy as np

def find_qa_dataset(hdf):
    """Try to find a QA-like dataset name"""
    possible_keywords = ['QA', 'quality', 'QC', 'reliability', 'Summary']
    for name in hdf.datasets().keys():
        lname = name.lower()
        if any(keyword in lname for keyword in possible_keywords):
            return name
    return None

def extract_cloud_mask(hdf_path, save_dir):
    """Extract cloud mask and save to a new HDF file"""
    hdf = SD(hdf_path, SDC.READ)

    qa_name = find_qa_dataset(hdf)
    if qa_name is None:
        print(f"No QA-related dataset found in {os.path.basename(hdf_path)}")
        return

    print(f" Using QA dataset: {qa_name}")
    qa_data = hdf.select(qa_name).get()

    # Example logic for MODIS QA: treat 0 as clear, others as cloudy
    cloud_mask = (qa_data != 0).astype(np.uint8)  # 1 = cloud, 0 = clear

    # Save cloud mask to new HDF
    new_filename = os.path.basename(hdf_path).replace('.hdf', '_cloudmask.hdf')
    new_hdf_path = os.path.join(save_dir, new_filename)

    new_hdf = SD(new_hdf_path, SDC.WRITE | SDC.CREATE)
    sds = new_hdf.create('CLOUD_MASK', SDC.UINT8, cloud_mask.shape)
    sds[:] = cloud_mask
    sds.endaccess()
    new_hdf.end()

# Folder containing MODIS .hdf files
data_folder = "/content/drive/My Drive/Sprints Final Project/Dateset"

#  Process all files
for filename in os.listdir(data_folder):
    if filename.endswith(".hdf"):
        file_path = os.path.join(data_folder, filename)
        extract_cloud_mask(file_path, data_folder)
```
